# xiachufang.com

This examples shows how to analyse infinite scroll from network trace.

```python
from IPython.core.display import Image
```

This website adopts an "infinite scroll" design for listing its items. More recipes will be loaded when the browser window hits the bottom of the page. One can notice that the URL in address bar stays the same. So the pagination URL can not be directly composed.

The solution is to use Chrome Developer Tool, the right side pane in following image. The tool can be enabled by `cmd`+`option`+`i`. Click into "Network" tab and filter network requests by "XHR" (stands for XmlHttpRequest; many ascynchronously data is loaded in this way). Now scroll and observe if there is any new network request.

```python
Image('xiachufang.com/unnamed.png')
```

Double-click the `?order=pop&page=5` request/ response. The browser will open this page in a new tab. One can see a complete 5th page loaded in the new tab. Observing the `page=5` parameter in the URL, one can change the number to find other pages.

```python
Image('xiachufang.com/unnamed-1.png')
```

## Code

```python
url_template = 'https://www.xiachufang.com/activity/site/?order=pop&page='
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
mypage = BeautifulSoup(requests.get(url_template + '1').text, 'lxml')
mypage.find_all('p', attrs={'class': 'name ellipsis red-font'})
```

```python
mypage = BeautifulSoup(requests.get(url_template + '2').text, 'lxml')
mypage.find_all('p', attrs={'class': 'name ellipsis red-font'})
```

## Further notes

Other pages may not be so directly. Most of the time infinite-scroll is implemented by loading XML/JSON files. You can still spell out the URL but needs to use different way to parse the file, e.g. `lxml` or `json`. The `BeautifulSoup` way we demoed in class works well with HTML.
