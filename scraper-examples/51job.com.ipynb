## 51job.com

This example demonstrates how to deal with non utf-8 encodings. UTF-8 encoding is the de facto standard but some websites still use other encoding methods. While `requests` does an excellent job in detecting the web page encoding, sometimes it still fails. You need to instruct the library the encoding to be used explicitly.

```python
from IPython.core.display import Image
```

One can identify the encoding from `meta` tag, in the `charset` attribute. See following screenshot:

```python
Image('51job.com/51job.com.png')
```

```python
import requests
```

```python
# search for frontend developing positions in Beijing and Hong Kong
url = 'https://search.51job.com/list/330000%252C010000,000000,0000,32,9,99,%25E5%2589%258D%25E7%25AB%25AF%25E5%25BC%2580%25E5%258F%2591,2,1.html?lang=c&stype=&postchannel=0000&workyear=99&cotype=99&degreefrom=99&jobterm=99&companysize=99&providesalary=99&lonlat=0%2C0&radius=-1&ord_field=0&confirmdate=9&fromType=&dibiaoid=0&address=&line=&specialarea=00&from=&welfare='
```

```python
r = requests.get(url)
```

```python
from pyquery import PyQuery
```

```python
pq = PyQuery(r.text)
```

```python
# Search for the company names
pq('span.t2 a').text()
```

One can see that there is something wrong with the encoding. We can tell the content encoding is `gbc` from the Chrome Developer Console. Let's fix as follows:

```python
r.encoding = 'gbk'
```

```python
pq = PyQuery(r.text)
pq('span.t2 a').text()
```
