# imdb.com

Key-points:

* Identify data points first (i.e. movie) and then identify variables (i.e. attributes)
* Save list-of-records type of data using pandas with ease

```python
import requests
import bs4
import pandas
```

```python
r = requests.get('http://www.imdb.com/list/ls058982125/')
```

```python
mypage = bs4.BeautifulSoup(r.text, 'lxml')
```

```python
mymovies = mypage.find_all('div', attrs={'class': 'lister-item-content'})
```

```python
#mymovies[0]
```

Note: The following parsing code made on April 22, 2018 is different from that made on Feb 23, 2018. Because the rating's HTML code has just changed. This is common for scraping. So be cautious about other's examples you can find online. Understanding the principle is very important for you to troubleshoot. The previous version can be found [here](https://github.com/hupili/python-for-data-and-media-communication/blob/master/w5-scraper-contd/imdb.ipynb) (`In [66]:`)

```python
movies = []
for mymovie in mymovies:
    mytitle = mymovie.find('h3').find('a').text
    myrating = mymovie.find(
        'div', attrs={'class': 'ipl-rating-star small'}
    ).find(
        'span', attrs={'class': 'ipl-rating-star__rating'}
    ).text
    myrating = float(myrating)
    myruntime = mymovie.find(
        'span', attrs={'class': 'runtime'}
    ).text
    movies.append([mytitle, myrating, myruntime])
```

```python
movies
```

```python
df = pandas.DataFrame(movies, columns=['title', 'rating', 'runtime'])
```

```python
df
```

```python
df.to_csv('imdb.com/imdb.com.csv')
```
