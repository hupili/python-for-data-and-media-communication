## Handle more pages

```python
import time
```

```python
pages = [
    ['filterdecade2010-2019', 'filteryear20102010-2018', 'filtermonth012010', 'filterissueNameJan31,2010January'],
    ['filterdecade2010-2019', 'filteryear20182010-2018', 'filtermonth102018', 'filterissueNameOct30,2018October'],
    ['filterdecade2010-2019', 'filteryear20182010-2018', 'filtermonth102018', 'filterissueNameOct29,2018October'],
    ['filterdecade2010-2019', 'filteryear20182010-2018', 'filtermonth102018', 'filterissueNameOct28,2018October']
]
```

```python
def scrape_one_page():
    items = browser.find_elements_by_class_name('resultItem')
    data = []
    for item in items:
        title = item.find_element_by_class_name('truncatedResultsTitle').text
        titleAuthorETC0 = item.find_elements_by_class_name('titleAuthorETC')[0].text
        titleAuthorETC1 = item.find_elements_by_class_name('titleAuthorETC')[1].text
        data.append((title, titleAuthorETC0, titleAuthorETC1, datestr))
    return data
```

```python
SLEEP_GAP = 5
all_data = []
for page in pages:
    try:
        browser.get(root_url)
        time.sleep(SLEEP_GAP)
        for link_id in page:
            browser.find_element_by_id(link_id).click()
            time.sleep(SLEEP_GAP)
        datestr = page[-1]
        data = scrape_one_page()
        all_data.extend(data)
    except Exception as e:
        print('Failed on ', page)
        print(e)
```

```python
df = pd.DataFrame(all_data, columns=['title', 'tae0', 'tae1', 'datestr'])
```
